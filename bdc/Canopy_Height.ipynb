```python
import PIL
import matplotlib.pyplot as plt
import numpy
```

```python
file = 'Raw Data/Forest_height_2019_AUS.tif'
```

```python
PIL.Image.MAX_IMAGE_PIXELS = 40257680016
im = PIL.Image.open(file)
width, height = im.size
```

```python
print(width, height)
```

```python
width_step_size = width / 1000
height_step_size = height / 1000
imarr = []
cord_arr = []

left = 0
top = 0
right = 0
bottom = 0

for i in range(int(width_step_size)): #width
    left += width_step_size
    right = left + width_step_size
    if right > width:
        right = width - left
    for h in range(int(height_step_size)): #height
        top += height_step_size
        bottom = top + height_step_size
        if bottom > height:
            bottom = height - top
        else:
            cord_arr.append((left, top, right, bottom))
            imarr.append(im.crop((left, top, right, bottom)))
```

```python
non_zero = set()
```

```python
for i in range(len(imarr)):
    local_w, local_h = imarr[i].size
    arr = numpy.array(imarr[i])
    for h in range(local_h):
        for w in range(local_w):
            if arr[h][w] != 0:
                non_zero.add(i)
                break
```

```python
n_z = list(non_zero)
cmp = []
for i in n_z:
   imarr[i].show()
```
